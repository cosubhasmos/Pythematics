# Visualizing Direction Field of Diffrential Equations


## What are direction fields?

Direction ﬁelds are valuable tools in studying the solutions of differential equations of the form,
$$\dfrac{dy}{dt} = f(t,y)$$
where $f$ is a given function of the two variables $t$ and $y$, sometimes referred to as the rate function. A direction ﬁeld for equations can be constructed by evaluating $f$ at each point of a rectangular grid. At each point of the grid, a short line
segment is drawn whose slope is the value of $f$ at that point. Thus each line segment is tangent to the graph of the solution passing through that point. A direction ﬁeld drawn on a fairly ﬁne grid gives a good picture of the overall behavior of solutions of a differential equation. Usually a grid consisting of a few hundred points is sufﬁcient. The construction of a direction ﬁeld is often a useful ﬁrst step in the investigation of a differential equation.

## Let's do it !

Firstly import necessary libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Now create a function named $f$ to define the ordinary differential equation $\dfrac{dy}{dx} = f(x, y)$. Consider a example,
$$ \dfrac{dy}{dx} = 3y - 1$$


```python

def f(x, y):
    return 3*y - 1
```

Define the range of $x$ and $y$ values. Be sure that equillibrium solution lies in between.

```python

y  = np.linspace(-5,5, 20)
x = np.linspace(-5,5, 20)
```

Create a grid of points


```python
X, Y = np.meshgrid(x, y)
```

Compute the direction of the vector field at each point


```python
U = 1  # Constant slope for x
V = f(X, Y)  # Slope for y given by the differential equation
```

Now let's plot the direction field.


```python
plt.figure(figsize=(8, 6))
plt.quiver(X, Y, U, V, scale=150, color='blue', alpha=0.5)
plt.xlabel('time',fontsize = 10,color='red')
plt.ylabel('y',fontsize = 10, color = 'red')
plt.title('Direction Field',fontsize=15,color = "purple")
plt.xticks(color="green",fontsize = 12)
plt.yticks(color="green",fontsize = 12)
plt.grid("science")
plt.show()
```

Now it's your turn to vary the equation and plot the direction fields similarly. Happy Coding !

Two observations are worth particular mention. First, in constructing a direction ﬁeld, we do not have to solve equation, but merely to evaluate the given function $f (x, y)$ many times. Thus direction ﬁelds can be readily constructed even for equations that may be quite difﬁcult to solve. Second, repeated evaluation of a given function is a task for which a computer is well suited, and you should usually use a computer to draw a direction ﬁeld.
